```python
import json
import glob
import os

# Update this pattern/path as needed
jsonl_files = glob.glob("/storage/home/dhameem.m.2025/City-of-Agents/CPG/raw/cpg_dataset_*.jsonl")

total_records = 0
total_json_errors = 0
total_error_field_count = 0
total_missing_graphml = 0

for path in jsonl_files:
    file_records = 0
    file_json_errors = 0
    file_error_field_count = 0
    file_missing_graphml = 0

    with open(path, "r", encoding="utf-8") as f:
        for line_no, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                continue

            try:
                rec = json.loads(line)
            except json.JSONDecodeError:
                file_json_errors += 1
                print(f"[JSON_DECODE_ERROR] {path}:{line_no}")
                continue

            file_records += 1

            # Script writes errors in rec["error"] when processing fails
            err = rec.get("error")
            if err not in (None, ""):
                file_error_field_count += 1
                print(f"[ERROR_FIELD] {path}:{line_no} idx={rec.get('idx')} error={err}")

            # Optional sanity check: successful rows should usually have graphml
            if not rec.get("graphml"):
                file_missing_graphml += 1

    total_records += file_records
    total_json_errors += file_json_errors
    total_error_field_count += file_error_field_count
    total_missing_graphml += file_missing_graphml

    print(f"\nFile: {path}")
    print(f"  records parsed: {file_records}")
    print(f"  malformed JSON lines: {file_json_errors}")
    print(f"  rows with error field: {file_error_field_count}")
    print(f"  rows missing graphml: {file_missing_graphml}")

print("\n=== Final Summary ===")
print(f"Files checked: {len(jsonl_files)}")
print(f"Total records parsed: {total_records}")
print(f"Total malformed JSON lines: {total_json_errors}")
print(f"Total rows with error field: {total_error_field_count}")
print(f"Total rows missing graphml: {total_missing_graphml}")

if total_json_errors == 0 and total_error_field_count == 0:
    print("PASS: No malformed JSON lines and no processing errors found.")
else:
    print("FAIL: Found malformed lines or rows with processing errors.")
```
